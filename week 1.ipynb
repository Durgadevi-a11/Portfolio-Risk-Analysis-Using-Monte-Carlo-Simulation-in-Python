```python
import yfinance as yf
import pandas as pd

stocks = ["RELIANCE.NS", "TCS.NS", "HDFCBANK.NS", "INFY.NS"]

df_list = []

for stock in stocks:
    temp = yf.download(stock, start="2020-01-01", progress=False)

    # Extract Close column properly from MultiIndex
    close_prices = temp.xs("Close", level="Price", axis=1)

    # Rename column to stock name
    close_prices.columns = [stock]

    df_list.append(close_prices)

# Combine all stocks
prices = pd.concat(df_list, axis=1)

prices.to_csv("raw_prices_data.csv")

print("Raw data downloaded successfully")
print(prices.head())
```

```python
df = prices.copy()

print("Missing before cleaning:\n", df.isnull().sum())

df = df[~df.index.duplicated(keep="first")]
df = df.ffill()
df.dropna(inplace=True)

df.to_csv("cleaned_prices_data.csv")

print("data cleaned Successfully")
```
